```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv("Model_Performance.csv", header=0)
plt.plot(df["epoch"], df["training_loss"], color = 'g', marker = ',',label = "Training Loss")
plt.plot(df["epoch"], df["training_acc"], color = 'b', marker = ',',label = "Training Accuracy")
plt.plot(df["epoch"], df["val_loss"], color = 'r', marker = ',',label = "Validation Loss")
plt.plot(df["epoch"], df["val_acc"], color = 'm', marker = ',',label = "Validation Accuracy")
plt.legend()
plt.xlabel("Epochs")
plt.show()
plt.savefig('model_performance.png')
```

```python
plt.plot(df["epoch"], df["training_loss"], color = 'g', marker = ',',label = "Training Loss")
plt.plot(df["epoch"], df["training_acc"], color = 'b', marker = ',',label = "Training Accuracy")
plt.plot(df["epoch"], df["val_loss"], color = 'r', marker = ',',label = "Validation Loss")
plt.plot(df["epoch"], df["val_acc"], color = 'm', marker = ',',label = "Validation Accuracy")
plt.ylim(top=1.2)
plt.ylim(bottom=0)
plt.legend(loc="lower right")
plt.xlabel("Epochs")
plt.figure(figsize=(12,8))
plt.show()

plt.savefig('model_performance_closeup.png')
```
